```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/igorkim/Documents/GitHub/hse_final_project/ml/0_data/expanded_data_with_OSM.csv', sep = ',')
```

```python
data.rename(columns = {'distance_to_bank_Сбербанк':'distance_to_sber_bank',
                       'distance_to_bank_ВТБ':'distance_to_vtb_bank',
                       'distance_to_bank_Альфа-Банк':'distance_to_alfa_bank',
                       'distance_to_bank_Росбанк':'distance_to_ros_bank',
                       'distance_to_atm_Сбербанк':'distance_to_sber_atm',
                       'distance_to_atm_ВТБ':'distance_to_vtb_atm',
                       'distance_to_atm_Альфа-Банк':'distance_to_alfa_atm',
                       'distance_to_atm_Росбанк':'distance_to_ros_atm'}, inplace = True)
```

```python
len(data)
```

```python
data.isnull().sum()
```

```python
#два пропущенных региона - это южный федеральный округ
data['regions'] = data['regions'].replace(np.nan, 'Southern Federal District')
```

```python
df_regions = pd.read_csv('/Users/igorkim/Documents/GitHub/hse_final_project/ml/0_data/regions_data.csv', sep = ';')
```

```python
data = data.merge(df_regions, how = 'inner', left_on = 'regions', right_on = 'Unnamed: 0').drop('Unnamed: 0', axis = 1)
```

```python
data.head()
```

```python
df_train= pd.read_csv('/Users/igorkim/Documents/GitHub/hse_final_project/ml/0_data/train.csv', sep = ',')
```

```python
df_train
```

```python
data = data.merge(df_train[['id','target']], how = 'inner',left_on = 'id', right_on = 'id')
```

```python
data = data.merge(data.groupby('cities').agg({'target': 'mean'}).reset_index().rename(columns={'target': 'avgC'}),
         how='left', on='cities')

data = data.merge(data.groupby('regions').agg({'target': 'mean'}).reset_index().rename(columns={'target': 'avgR'}),
         how='left', on='regions')

data = data.merge(data.groupby('states').agg({'target': 'mean'}).reset_index().rename(columns={'target': 'avgS'}),
         how='left', on='states')

data = data.merge(data.groupby('atm_group').agg({'target': 'mean'}).reset_index().rename(columns={'target': 'avgA'}),
         how='left', on='atm_group')
```

```python
data['capital'] = np.where((data['cities'] == 'Moscow') | (data['cities'] == 'Saint Petersburg'), 1, 0)
```

```python
data.to_csv('/Users/igorkim/Documents/GitHub/hse_final_project/ml/0_data/expanded_data_with_OSM_v3.csv')
```
